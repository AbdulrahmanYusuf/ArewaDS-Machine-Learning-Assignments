**Visualizing Quantities**

Let ispect data or dataset to understand the items involve in the data first

```python
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('../../ArewaDS-Machine-Learning-Assignments/week3/birds.csv')
birds.tail()
```

**Build a line plot about bird bodyweight**

Let plot some of the numeric data using a basic line plot for the datasets. . Assuming we wanted a view of the maximum bodyweight for these interesting birds. 


```python
plt.title('Fig 1: Max Body Weight in gramms')
bodyeight = birds['MaxBodyMass'] 
bodyeight.plot()
```

Let add labels to the x-axis to show what kind of birds are in question:

```python
plt.title('Fig 2: Max Body Weight in gramms')
plt.ylabel('BodyWeight (g)')
plt.xlabel('Birds')
plt.xticks(rotation=45)
x = birds['Name'] 
y = birds['MaxBodyMass']

plt.plot(x, y)

plt.show()
```

Let us bringout more information for clarity

```python
plt.title('Fig 3: Max BodyWeight in gramms')
plt.ylabel('BodyWeight (g)')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxBodyMass'][i]
    plt.plot(x, y, 'bo')
    if birds['MaxBodyMass'][i] > 10000:
        plt.text(x, y * (1 - 0.05), birds['Name'][i], fontsize=10)
    
plt.show()
```

Let remove some outliers from the data through means of filtering

```python
plt.title('Fig 4: Max BodyWeight in gramms')
plt.ylabel('BodyWeight (g)')
plt.xlabel('Birds')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)
for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxBodyMass'][i]
    if birds['Name'][i] not in ['Wild turkey', 'Trumpeter swan', 'American white pelican', 'Mute swan']:
        plt.plot(x, y, 'bo')
plt.show()
```
