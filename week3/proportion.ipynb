
**Visualising Proportion**

Let use a nature-focused dataset to visualize proportions. Let inspect to understand about the dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
mushroom = pd.read_csv('../../ArewaDS-Machine-Learning-Assignments/week3/mushrooms.csv')
mushroom.tail()
```

To know the columns and the data types of the column

```python
col = mushroom.select_dtypes(['object']).columns
col
```

To know the number of entry in each column

```python
print(mushroom.select_dtypes(['object']).count())
```

Let consider a column *odour* to make a onversion and draw a pie-chart

```python
smellclass=mushroom.groupby(['odor']).count()
smellclass
```

```python
lebel=['Almnd', 'Anise', 'Creosote', 'Fishy', 'Foul', 'Musty', 'None', 'Pungent', 'Spicy']
plt.pie(smellclass['population'], labels=lebel, autopct='%.1f %%')
plt.title('Fig. 1: Mushrooms Classification Based on Odour')
plt.show()
```

**Donuts**

Is just a sub-class of pie-chart

Take a look at the various population of different type of mushrooms 

```python
Population=mushroom.groupby(['population']).count()
Population
```

Here, we grouped that databy population. There are five listed different population, so use those as labels for the donut chart:

```python
labels=['Abundant','Clustered','Numerous','Scattered','Several','Solitary']

plt.pie(Population['odor'], labels=labels,
        autopct='%1.1f%%', pctdistance=0.85)

  
center_circle = plt.Circle((0, 0), 0.30, fc='white')
fig = plt.gcf()

fig.gca().add_artist(center_circle)
  
plt.title('Fig. 2: Mushroom Population Odor')
  
plt.show()
```

This code insruct to draw a chart and a center circle, then adds that center circle in the chart. you can change the width of the center circle by changing 0.30 to another value.

Donut charts can be tweaked in several ways to change the labels. The labels in particular can be highlighted for readability. you can learn more in the documentation.

```python
labels=['Almnd', 'Anise', 'Creosote', 'Fishy', 'Foul', 'Musty', 'None', 'Pungent', 'Spicy']


#df = pd.DataFrame({'labels': labels, 'sizes': odor})

# Create a figure and a subplot
fig, ax = plt.subplots()

# Create a scatter plot of the data
ax.scatter(smellclass['class'], smellclass.index, color='red')

# Draw vertical lines from the scatter points to the x-axis
ax.vlines(smellclass['class'], ymin=smellclass.index, ymax=-1, color='red')

# Set the y-axis ticks and labels
ax.set_yticks(smellclass.index)
ax.set_yticklabels(smellclass['class'])

# Set the axis labels
ax.set_xlabel('Class')
ax.set_ylabel('Labels')

# Set the title
plt.title('Fig. 3: Mushroom  Odor')

# Display the chart
plt.show()
```

Waffle

```python
from pywaffle import Waffle
capcolor=mushroom.groupby(['cap-color']).count()  
data ={'color': ['brown', 'buff', 'cinnamon', 'green', 'pink', 'purple', 'red', 'white', 'yellow'],
    'amount': capcolor['class']
     }
  
df = pd.DataFrame(data)
  
fig = plt.figure(
    FigureClass = Waffle,
    rows = 100,
    values = df.amount,
    labels = list(df.color),
    figsize = (30,30),
    colors=["brown", "tan", "maroon", "green", "pink", "purple", "red", "whitesmoke", "yellow"]
)
```
