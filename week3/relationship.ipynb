```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
honey = pd.read_csv('../../ArewaDS-Machine-Learning-Assignments/week3/honey.csv')
honey.tail()
```

We can now understand the interesting available columns of the honey production in U.S.A. The data having total of 626 row with eight columns

```python
honey.shape
```

Let create a basic scatterplot to depict the relationship between the yield per colony and its U.S. state of origin. let make the y axis tall enough to display all the available states:

```python
sns.relplot(x="yieldpercol", y="state", data=honey, height=15, aspect=.5);
plt.title('Fig 1: Yield per Colony based on US state')
```

Now,  some color scheme are used to show same data as above. That is, to show how the yiels per colony over the years. You can do this by adding a 'hue' parameter to show the change, year over year:

```python
sns.relplot(x="yieldpercol", y="state", hue="year", palette="YlOrBr", data=honey, height=15, aspect=.5);
plt.title('Fig 2: Yield per Colony based on US state and use color to show changes over years')
```

Let observe the date distribution from state of Florida

```python
honey.loc[honey['state']=='FL']
```

Let get the data in more visualised form by increasing the size of the datscatterd points year over over

```python
sns.relplot(x="yieldpercol", y="state", size="year", data=honey, height=15, aspect=.5);
plt.title('Fig 3: Yield per Colony Based on State')
```

LineChart

Observe the chart below there is fall in the yield per colony over years. There is a  graet fall in yield per colony between 2008 and 2010

Facet grids

Facet grids allows to take one facet of a dataset (let choose 'year' facet). Seaborn enables to make a plot for each of those facets of the chosen x and y coordinates for clear visual comparison. 

```python
sns.relplot(
    data=honey, 
    x="prodvalue", y="yieldpercol",
    col="year", 
    col_wrap=3,
    kind="line")
plt.title('Fig 6: Yield per Colony Based on State')
```

In this visualization, you can compare the product values and yield per colony year over year, side by side with a wrap set at 3 for the columns:

Dual-line Plots

This type of chart allows superimposing of two lineplots on top of each other, using Seaborn's 'despine' to remove their top and right spines, and using ax.twinx derived from Matplotlib. Twinx allows a chart to share the x axis and display two y axes. So, display the product values and yield per colony, superimposed:

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=honey['year'], y=honey['prodvalue'], data=honey, 
                        label = 'Product value', legend=False)
sns.despine()
plt.ylabel('ProductValue')
plt.title('Honey Production Year over Year');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=honey['year'], y=honey['yieldpercol'], ax=ax2, color="r", 
                         label ='Yield per colony', legend=False) 
sns.despine(right=False)
plt.ylabel('colony yield')
ax.figure.legend();
```
