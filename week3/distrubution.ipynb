```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
birds = pd.read_csv('../../ArewaDS-Machine-Learning-Assignments/week3/birds.csv')
honey = pd.read_csv('../../ArewaDS-Machine-Learning-Assignments/week3/honey.csv')
birds.tail()
```

Box Plots

```python
# create a boxplot from our dataset honey
plt.boxplot(honey['yieldpercol'])
plt.title('Fig 1: Box plot for yield per colony')
plt.show()
```

Violin plot

```python

plt.violinplot(honey['yieldpercol'])
plt.title('Fig 2: violin plot for yield per colony')
plt.show()
```

```python
birds.plot(kind='scatter',x='MaxBodyMass',y='Family',figsize=(8,14))

plt.title('Fig 3: Max Bodyweight per family')
plt.ylabel('Family')
plt.xlabel('Max Weight')

plt.show()
```

```python
birds['MaxLength'].plot(kind = 'hist', bins = 10, figsize = (12,12))
plt.title('Fig 5: Max of birds per family')
plt.show()
```

```python
birds['MaxLength'].plot(kind = 'hist', bins = 30, figsize = (12,12))
plt.title('Fig 6: Max of batds per family')
plt.show()
```

```python
filteredBirds = birds[(birds['MaxLength'] > 3) & (birds['MaxLength'] < 40)]      
filteredBirds['MaxLength'].plot(kind = 'hist',bins = 40,figsize = (12,12))
plt.title('Fig 7: bird maximum length between 1 and 60 ')
plt.show()
```

```python
x = filteredBirds['MaxLength']
y = filteredBirds['MaxBodyMass']

fig, ax = plt.subplots(tight_layout=False)
hist = ax.hist2d(x, y)
plt.title('Fig 8: Max of birds per family')
```

```python
x1 = filteredBirds.loc[filteredBirds.ConservationStatus=='EX', 'MinLength']
x2 = filteredBirds.loc[filteredBirds.ConservationStatus=='CR', 'MinLength']
x3 = filteredBirds.loc[filteredBirds.ConservationStatus=='EN', 'MinLength']
x4 = filteredBirds.loc[filteredBirds.ConservationStatus=='NT', 'MinLength']
x5 = filteredBirds.loc[filteredBirds.ConservationStatus=='VU', 'MinLength']
x6 = filteredBirds.loc[filteredBirds.ConservationStatus=='LC', 'MinLength']

kwargs = dict(alpha=0.5, bins=20)

plt.hist(x1, **kwargs, color='red', label='Extinct')
plt.hist(x2, **kwargs, color='orange', label='Critically Endangered')
plt.hist(x3, **kwargs, color='yellow', label='Endangered')
plt.hist(x4, **kwargs, color='green', label='Near Threatened')
plt.hist(x5, **kwargs, color='blue', label='Vulnerable')
plt.hist(x6, **kwargs, color='gray', label='Least Concern')

plt.gca().set(title='Fig 9: Conservation Status', ylabel='Min Wingspan')
plt.legend();
```
