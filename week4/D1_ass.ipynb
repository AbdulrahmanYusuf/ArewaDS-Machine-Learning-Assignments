```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
# Load the linnerud dataset
linnerud_X, linnerud_y = datasets.load_linnerud(return_X_y=True, as_frame=False)
```

```python
plt.violinplot(y=linnerud_X, x=linnerud_y)
plt.show
```

```python
# Use only one feature
linnerud_X= linnerud_X[:, np.newaxis, 0]
linnerud_y= linnerud_y[:, np.newaxis, 0]
```

```python
#spliting the data for train and test
linnerud_X_train, linnerud_X_test, linnerud_y_train, linnerud_y_test = train_test_split(linnerud_X, linnerud_y, test_size=0.25)

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(linnerud_X_train, linnerud_y_train)

# Make predictions using the testing set
linnerud_y_pred = regr.predict(linnerud_X_test)

# The coefficients
print("Coefficients: \n", regr.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(linnerud_y_test, linnerud_y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(linnerud_y_test, linnerud_y_pred))

# Plot outputs
plt.scatter(linnerud_X_test, linnerud_y_test, color="black")
plt.plot(linnerud_y_test, linnerud_y_pred, color="blue", linewidth=2)

plt.xticks(())
plt.yticks(())

plt.xlabel('size measures')
plt.ylabel('size Progression')
plt.title('A Graph Plot Showing linnerud Progression Against measures')
plt.show()
```
