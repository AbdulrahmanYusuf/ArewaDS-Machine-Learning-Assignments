```python
#CELL-1: Importing some libraries

import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
from sklearn.metrics import mean_squared_error, r2_score
```

importing matplotlib, numpy and you are importing datasets, linear_model and model_selection from sklearn. model_selection is used for splitting data into training and test sets.

```python
#CELL-2: loading the datasets and assigned to variable x, and y fairs in form of dataframe

X, y = datasets.load_diabetes(return_X_y=True, as_frame=True)
print(X.shape) # to know the datasets diamension
print(X)   # to print the first column of the datasets
```

You can notice from the above, we have added the parameter *as_frame* in order have clear picture of the data in form of rows and columns

```python
#CELL-3: loading the datasets in form of array

X, y = datasets.load_diabetes(return_X_y=True)
```

Selecting the portion of the datasets to plot with. third column is selected in this case

```python
#CELL-4: We selected the third column of the from the dataset

X = X[:, np.newaxis, 2]
X
```

Our dataset can be splited into train and test. Preparably, train portion should alway be greater than the test propotion

```python
#CELL-5: splittig datasets into train and test by 75% and 25% respectively

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

Let's us load up the linear regression model for training with our X and y training sets:

```python
#CELL-6: loading for training and fitting

model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

We then create a prediction from the test data, using the function predict()

```python
#CELL-7 prediction creation
y_pred = model.predict(X_test)
```

Let use power of matplotlib to draw the linear regression graph

```python
#CELL-8

plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```

```python
# The coefficients
print("Coefficients: \n", model.coef_)
# The mean squared error
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))
```
